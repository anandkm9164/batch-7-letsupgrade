```python
import math

def cone(radius, height):
    l = math.pi * radius * (math.sqrt(radius * radius + height * height))

    SA = math.pi * radius * radius

    valume = math.pi * radius * radius * (height / 3)

    print("\n side of a cone = %.2f" % l)
    print("\n surface area of a cone = %.2f" %SA)
    print("\n valume o of a cone = %.2f" %valume)

cone(6, 10)

```
